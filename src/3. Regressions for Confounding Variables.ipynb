# Confounder Identification

### Helper Function

```python
def id_confounder(data, treatment, outcome, list_confounders):
    # df to store model results
    results_df = pd.DataFrame(columns=["Outcome", "Treatment", "Treatment Slope", "P-Value Treatment", "R-Sq", "Potential_Confounder", "P-Value with Confounder", "F-Stat"])

    # regress outcome on treatment without any confounders first
    X = data[treatment]
    y = data[outcome]
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    
    coefficients = results.params
    treatment_slope = round(coefficients[treatment], 2)
    p_val_without_conf = ["Significant" if results.pvalues.iloc[1] < 0.05 else "Not Significant"][0]
    r_squared_without_conf = round(results.rsquared, 2)
    f_stat = ["Significant" if results.f_pvalue < 0.05 else "Not Significant"][0]
    row = [outcome, treatment, treatment_slope, p_val_without_conf, r_squared_without_conf, 
           "-", "-", f_stat]
    
    results_df.loc[len(results_df)] = row

    # regress outcome on treatment with confounder
    for c in list_confounders:
       #  print(c)
        X = data[[treatment, c]]
        X = sm.add_constant(X)
        model_with_confounder = sm.OLS(y, X)
        results_with_confounder = model_with_confounder.fit()

        coefficients_with_conf = results_with_confounder.params
        treatment_slope = round(coefficients_with_conf[treatment], 2)
        p_val_treatment = ["Significant" if results_with_confounder.pvalues.iloc[1] < 0.05 else "Not Significant"][0]
        p_val_confounder = ["Significant" if results_with_confounder.pvalues.iloc[2] < 0.05 else "Not Significant"][0]
        r_squared_with_confounder = round(results_with_confounder.rsquared, 2)
        f_stat = ["Significant" if results_with_confounder.f_pvalue < 0.05 else "Not Significant"][0]

        row = [outcome, treatment, treatment_slope, p_val_treatment, r_squared_with_confounder, 
               c, p_val_confounder, f_stat]

        results_df.loc[len(results_df)] = row


    return results_df


```

## For "units_sold" ~ "overall_rating"

```python
treatment = "overall_rating"
outcome = "units_sold"

potential_numerical_confounders = data.select_dtypes(["float64", "int64"]).columns.tolist()
# remove treatment and outcome from confounders
potential_numerical_confounders.remove("overall_rating")
potential_numerical_confounders.remove("units_sold")

results_df = id_confounder(data, treatment, outcome, potential_numerical_confounders)

results_df
```
